```python
# matplotlib contains plotting libraries for python.

# "matplotlib.pyplot" is collection of command style function that make it to wwork like MATLAB.

# pyplot function used to create figure,create plotting area in figure,plot lines in plotting area,assign labels to plot.

import matplotlib.pyplot as pypl
```

```python
# Line graph using plot() -

pypl.plot([2,4,6,3,5])     # plot graph using values passed to it as y-axis & their index no. is used as x-axis.

pypl.ylabel("value")       # assign label to y-axis

pypl.xlabel("Index")       # assign label to x-axis

pypl.title("pyplot graph")     #  add title to graph

pypl.show()      # shows graph with label & title
```

```python
# Line graph with grid() -

pypl.plot([1,2,3,4],[1,4,9,16])    # passing to list 1st for x-axis & 2nd for y-axis

pypl.ylabel("squares")

pypl.xlabel("numbers")

pypl.grid()    # apply grid to graph for better understanding

pypl.show()
```

```python
# format string - indicate color and line-type of plot

pypl.plot([1,2,3,4],[1,4,9,16],'ro')     # here r means red-color & o means dot symbol

pypl.grid()    

pypl.show()
```

```python
# Multiple lined graph : 

import numpy as np

p=np.arange(0.,5.,0.2)    # create array ranges from 0 to 5 with interval of 0.2

pypl.plot(p,p**2,"b--",label="^2")     # exponent of 2 , "b--":print blue line in dash symbol , power of 2

pypl.plot(p,p**2.2,"rs",label="^2.2")  # exponent of 2.2 , "rs":print red line in square symbol , power of 2.2 

pypl.plot(p,p**2.5,"g^",label="^2.5")  # exponent of 2.5 , "g^":print green line in triangle symbol , power of 2.5

pypl.grid()

pypl.legend()     # used to specify which line belongs what data , values in labels will print legends

pypl.show()
```

```python
# Controlling line properties -

# A) using keyword args :

x=[2,4,6,8]

y=[10,20,30,40]

pypl.plot(x,y,linewidth=5.0)     # linewidth increases the size of line

pypl.grid()

pypl.show()
```

```python
# B) using setp() : set-property

x1=[1,2,3,4]

y1=[1,4,9,16]    # square of x1

x2=[1,2,3,4]

y2=[2,4,6,8]     # double of x2

lines=pypl.plot(x1,y1,x2,y2)

pypl.setp(lines[0],color='r',linewidth=2.0)     # python style keyword-args ; lines[0]=x1,y1   

pypl.setp(lines[1],'color','g','linewidth',2.0)    # MATLAB style string-value pair ; lines[1]=x2,y2 

pypl.grid()
```

```python
# Working with multiple figures & axis - which breaks single figure in two parts

# subplot() shows number of rows & column.

# figure_num ranges from 1 to number_of_rows * number_of_columns

import numpy as np

def f(t):
    
    return np.exp(-t)*np.cos(2*np.pi*t)


f1=np.arange(0.0,5.0,0.1)

f2=np.arange(0.0,5.0,0.02)

pypl.figure(1)      # create space for figure 1

pypl.subplot(211)   # here 2=no_of_row,1=n0_of_column,1=figure_num

pypl.grid()

pypl.plot(f1,f(f1),'b-')


pypl.subplot(212)   # here 2=no_of_row,1=n0_of_column,2=figure_num

pypl.grid()

pypl.plot(f2,np.cos(2*np.pi*f2),'r--')


pypl.show()
```

```python
pypl.figure(1)              # create space for figure 1 

fig,axes=pypl.subplots(nrows=4,ncols=4)    # used to fit subplot in figure

fig.tight_layout()

pypl.subplot(211)          # divide the figure in subplot 

pypl.plot([1,2,3])         # figure 1 : subplot 1 

pypl.title('Figure 1')


pypl.subplot(212)          # figure 1 : subplot 2

pypl.plot([4,5,6])         

pypl.title('Subplot 2')


pypl.figure(2)             # figure 2

fig,axes=pypl.subplots(nrows=1,ncols=1)

fig.tight_layout()

pypl.plot([7,8,9])         # create subplot(111) by default

pypl.title('Figure 2')


#pypl.figure(1)          # adding title to first-subplot of figure 1

#pypl.subplot(211)

#pypl.title('Subplot 1')


pypl.show()    # common to all plots

```
